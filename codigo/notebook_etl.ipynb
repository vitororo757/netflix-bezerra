# Notebook - ETL
#### O objetivo deste notebook é juntar e padronizar os dados dos .csvs necessários para a análise, para que possam ser consumidos no Power BI

## 0 - Importações

```python
import pandas as pd
```

## 1 - Unindo os dados de todos os integrantes do grupo em um único DataFrame

```python
my_list = [
    '../dados/GabrielKoji/MyListKoji.csv',
    '../dados/JoãoVargas/MyListJoao.csv',
    '../dados/RafaelCruz/MyListRafa.csv',
    '../dados/RaissaCasale/MyListRaissa.csv',
    '../dados/RyanLionel/MyListRyan.csv'
]

df_mylist = pd.concat([pd.read_csv(f) for f in my_list], ignore_index=True)

display(df_mylist.head(5))
```

```python
viewing = [
    '../dados/GabrielKoji/ViewingActivityKoji.csv',
    '../dados/JoãoVargas/ViewingActivityJoao.csv',
    '../dados/RafaelCruz/ViewingActivityRafa.csv',
    '../dados/RaissaCasale/ViewingActivityRaissa.csv',
    '../dados/RyanLionel/ViewingActivityRyan.csv'
]

df_viewing = pd.concat([pd.read_csv(f) for f in viewing], ignore_index=True)

display(df_viewing)
```

## 2 - Exportando os DataFrames transformados para a pasta dados-transformados

```python
df_mylist.to_csv('../dados-transformados/MyList.csv', index=False)
df_viewing.to_csv('../dados-transformados/Viewing.csv', index=False)
```
